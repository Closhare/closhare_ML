```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import os, re, requests, pandas as pd
from time import sleep
from selenium.webdriver.chrome.service import Service as ChromeService 
from bs4 import BeautifulSoup
import os, re, requests, pandas as pd

CHROMEDRIVER_PATH = r"C:\Users\wendy\Desktop\closhare\chromedriver-win64\chromedriver-win64\chromedriver.exe"

def init_driver(headless=True):
    options = Options()
    if headless:
        options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--window-size=1920,1080')
    options.add_argument('--disable-blink-features=AutomationControlled')
    service = ChromeService(executable_path=CHROMEDRIVER_PATH)
    return webdriver.Chrome(service=service, options=options)

def scroll_down(driver, scrolls=8, delay=0.25):
    """페이지 끝까지 천천히 스크롤하면서 lazy-loading 유도"""
    for _ in range(scrolls):
        driver.execute_script("window.scrollBy(0, window.innerHeight);")
        sleep(delay)

def fetch_html_by_selenium(url: str) -> str:
    driver = init_driver()
    driver.get(url)
    scroll_down(driver)
    html = driver.page_source
    driver.quit()
    return html

def parse_img_srcs_from_file(html_path: str) -> list[str]:
    with open(html_path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    imgs = soup.find_all("img")
    srcs = []
    for img in imgs:
        src = img.get("src") or img.get("data-src")
        if src and src.startswith("https://media.bunjang.co.kr/product/"):
            # 쿼리스트링 제거
            clean_src = src.split("?")[0]
            srcs.append(clean_src)

    return list(dict.fromkeys(srcs))  # 중복 제거

def download_all(srcs: list[str], page):
    for i, src in enumerate(srcs, 1):
        filename = f"{page}_{i:03}.jpg"
        path = os.path.join(SAVE_DIR, filename)
        try:
            r = requests.get(src, timeout=10)
            if r.status_code == 200:
                with open(path, "wb") as f:
                    f.write(r.content)
                print(f"✅ 저장: {filename}")
            else:
                print(f"⚠️ 실패: {src} [HTTP {r.status_code}]")
        except Exception as e:
            print(f"⚠️ 예외: {src} → {e}")
```

```python
SAVE_DIR = "images_악세사리"
os.makedirs(SAVE_DIR, exist_ok=True)

for page in range(1,6):
    URL = f"https://m.bunjang.co.kr/categories/400?page={page}&req_ref=popular_category"
    HTML_PATH = "fetched_bunjang.html"
    
    html = fetch_html_by_selenium(URL)
    
    # ⬇️ HTML 일부 확인
    with open("fetched_bunjang.html", "w", encoding="utf-8") as f:
        f.write(html)
    print("✅ HTML 저장 완료 → fetched_bunjang.html")
    
    imgs = parse_img_srcs_from_file(HTML_PATH)
    print(f"🔍 발견된 이미지: {len(imgs)}개")
    download_all(imgs, page)

    print("✅ 다운로드 완료 및 로그 저장!")
```
